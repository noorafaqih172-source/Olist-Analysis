```python
# Actual weight in kg
items['actual_kg'] = items['product_weight_g'] / 1000

# Volumetric weight
items['volumetric_kg'] = (
    items['product_length_cm'] *
    items['product_width_cm'] *
    items['product_height_cm']
) / 5000

# Chargeable weight = max of the two for each row
items['Chargeable_Weight'] = items[['actual_kg', 'volumetric_kg']].max(axis=1)
```

```python
# Calculate the average weight for 'bed_bath_table', excluding zeros
avg_weight = items.loc[
    (items['product_category_name'] == 'bed_bath_table') & (items['product_weight_g'] > 0),
    'product_weight_g'
].mean()

# Replace zeros with that average only for that category
items.loc[
    (items['product_category_name'] == 'bed_bath_table') & (items['product_weight_g'] == 0),
    'product_weight_g'
] = avg_weight
```
